# Personal Expense Analytics\n\nThis notebook reads the processed CSV files generated by the ETL pipeline and performs quick portfolio-ready analysis.

```python
from pathlib import Path\nimport pandas as pd\n\nproject_root = Path('..')\ndata_dir = project_root / 'data' / 'processed'\n\ntransactions = pd.read_csv(data_dir / 'transactions_clean.csv', parse_dates=['transaction_date'])\nmonthly = pd.read_csv(data_dir / 'monthly_summary.csv')\nrecurring = pd.read_csv(data_dir / 'recurring_payments.csv')\nmerchants = pd.read_csv(data_dir / 'merchant_summary.csv')\n\ntransactions.head()
```

```python
summary = {\n    'total_income': transactions['deposit_amount'].sum(),\n    'total_expenses': transactions['withdrawal_amount'].sum(),\n    'net_savings': transactions['deposit_amount'].sum() - transactions['withdrawal_amount'].sum(),\n    'transaction_count': len(transactions),\n}\nsummary
```

```python
category_spend = (\n    transactions.query('withdrawal_amount > 0')\n    .groupby('category', as_index=False)['withdrawal_amount']\n    .sum()\n    .sort_values('withdrawal_amount', ascending=False)\n)\ncategory_spend
```
